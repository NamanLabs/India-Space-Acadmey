```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
gdf = gpd.read_file(r"C:\Users\naman\Desktop\INTERNSHIP\INDIA SPACE ACADMEY\23-5-26\SHP FILE FORMED"
)
gdf.head()
```

```python
gdf.plot(
    figsize=(10,10),
    edgecolor='black'
)

plt.title("India District Boundaries")

plt.show()
```

```python
gdf.plot(
    column="mean_value",
    cmap="YlOrRd",
    legend=True,
    figsize=(10,10),
    edgecolor="black",
    linewidth=0.3
)

plt.title("District-wise Mean Raster Values")

plt.show()
```

```python
ts = pd.DataFrame({
    "Year": [2022, 2023, 2024, 2025],
    "Mean_Value": [10.5, 11.2, 12.4, 13.1]
})
ts
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    ts["Year"].astype(str),
    ts["Mean_Value"],
    marker="o",
    linewidth=2
)

plt.title("Time-Series Change in Mean Raster Value")

plt.xlabel("Year")

plt.ylabel("Mean Raster Value")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    ts["Year"].astype(str),
    ts["Mean_Value"]
)

plt.title("Year-wise Mean Raster Values")

plt.xlabel("Year")

plt.ylabel("Mean Raster Value")

plt.show()
```

```python
ts["Another_Value"] = [9.8, 10.6, 11.5, 12.0]

plt.figure(figsize=(7,5))

plt.scatter(
    ts["Mean_Value"],
    ts["Another_Value"]
)

plt.title("Relationship Between Two Variables")

plt.xlabel("Mean Value")

plt.ylabel("Another Variable")

plt.grid(True)

plt.show()
```

```python
correlation = ts["Mean_Value"].corr(
    ts["Another_Value"]
)
print("Correlation value:")

print(correlation)
```
